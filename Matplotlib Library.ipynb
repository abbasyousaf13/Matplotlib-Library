## Line plots

```python
x = np.linspace(0,10,1000)
# using line plot
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
x = np.linspace(0,10,1000)
# using line plot
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show() # In Jupyter one is not required to write plt.show() but in others IDE it is must
```

```python
# subplot can be done by state based programming
# subplot divides in pieces
plt.Figure()

plt.subplot(2,1,1)       #   2/1  ==>   2 = rows   &   1 = column   &  other 1 = sequence number
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
# subplot can alos be done by object oriented programming
fig, ax = plt.subplots(2)     # here fig will be used to store figure while ax will be used to store axis as 1 and 2 respectivley

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
fig.savefig("linePlot.png")
```

```python
plt.style.use('seaborn-whitegrid')    # using matplotlib style sheets
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x))
plt.show()
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = '#000', linestyle = 'solid', label = "sin(x)")   # black color hash code
plt.plot(x, np.sin(x-1), color = 'g', linestyle = 'dashed', label = "sin(x-1)")
plt.plot(x, np.sin(x-2), color = 'r', linestyle = 'dotted', label = "sin(x-2)")
plt.plot(x, np.sin(x-3), color = '#FFBF00', linestyle = 'dashdot', label = "sin(x-3)")
plt.plot(x, np.sin(x-4), color = 'y', linestyle = '-.', label = "sin(x-4)")
plt.plot(x, np.sin(x-5), color = '#40E0D0', linestyle = ':', label = "sin(x-5)")
plt.plot(x, np.sin(x-6), ":g", label = "sin(x-6)")        # this is combination of color and linestyle. it can also be sued like this
plt.plot(x, np.sin(x-6), "-.r", label = "sin(x-7)")

# labeling can be used by using title function
plt.title("Sin Graph")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.show()
```
